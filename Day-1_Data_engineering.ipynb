#  Data Engineering:
Data Engineering involves building systems and pipelines to collect, store, and analyze large volumes of data for downstream applications like analytics, ML, and reporting.


**Data**: Raw, unorganized, and unanalyzed facts, figures, or symbols.  
**Information**: Data that has been processed, organized, structured, or contextualized. 


**Data engineering Lifecycle**:  
Data collection >> Ingestion >> Processing >> Storage >> Analysis >> Consumption.

**Data Sources**:  
Databases, APIs, Files, Event Streams(Kafka, IoT Devices), Web Scraping(Scrapy, BeautifulSoup)

**Types of data:**  
1) Structured Data - Organised in a format (tables)
2) Unstructured - Unorganised data(images, audios)
3) Semi Structured - Partially organised data( Jason data)


**Data Ingestion** :  
1) Batch Ingestion : collecting data over a period of time and loading it as a group into a central system, like a database or data lake.(Periodic ingestion) (eg: Daily logs) 
2) Real-Time Ingestion :  the continuous process of collecting and moving data into a system the exact moment it is created. (Streaming data, eg: transactions)

**ETL and ELT for Ingestion:**    
ETL and ELT are two approaches to building data pipelines, primarily differing in when data is transformed. ETL transforms data before loading it into a target system. ELT loads raw data into a destination first, and then transforms it.  

ETL(Extract, Transform, Load):  
Data is pulled from source systems (like CRMs or ERPs), cleaned and structured in a staging area, and then loaded into the target database.  
Best used for: Smaller datasets, structured data, strict data governance requirements, or systems requiring compliance with privacy regulations before data enters a warehouse.  

ELT (Extract, Load, Transform):  
It extracts data from various sources and loads it directly into a data warehouse in its raw format. Data is then transformed as needed using the warehouse's own computing power.  
Best used for: Large volumes of data, unstructured/semi-structured data (like documents or images), and real-time analytics where quick data ingestion is a priority. 

**Data Pipelines**  
Pipeline Design: - Modular and reusable components - Error handling, retries

Orchestration: - Using Airflow, Prefect, or Dagster to schedule and manage tasks  
Data pipeline orchestration is the automated management, scheduling, and monitoring of data workflows. It coordinates complex tasks—such as data ingestion, cleaning, transformation, and delivery—ensuring steps run in the correct order, dependencies are respected, and failures are handled without manual intervention.

CI/CD for Pipelines: - Using GitHub Actions, Jenkins for deployment


**Data Governance & Security**  
• Data Lineage: the process of tracking data from its origin, through all transformations and movements, to its final destination  
• Cataloging : the systematic process of creating metadata to describe, organize, and index information resources (like books, documents, or digital files).  
• Masking & Encryption:  
•Masking: Hide sensitive data like phone numbers or credit card info.  
Example: Show ****1234 instead of full card number  
•Encryption: Lock the data so only authorized users can open/read it.  
Like turning the data into a secret code  



**Data Visualization & Reporting**  
Power BI:  
Data Import & Transformation: Connect to data sources, clean and transform data, remove unnecessary information, and choose the appropriate storage mode (Import or DirectQuery).  
Data Modeling: Build relationships between tables using a star schema, define proper relationships, and use a date table for accurate analysis.  
DAX Measures: Create reusable calculations with DAX measures, leverage filter context, and use functions like CALCULATE and time intelligence for dynamic insights.  
Dashboard Design: Design clear, visually appealing dashboards with KPIs, effective color usage, and minimal clutter for better readability.  



Tableau:  
Data Connections: Connect Tableau to multiple data sources using live connections or data extracts.  
Custom Calculations: Create calculated fields and formulas to analyze and transform data.  
Interactive Dashboards: Build dynamic dashboards with filters, parameters, and drill-down capabilities.  
Data Visualization: Present insights using charts, maps, graphs, and KPIs for effective decision-making.
