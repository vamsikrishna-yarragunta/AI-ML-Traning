# Functions In Python  
A function is a block of code which we can run multiple times by calling it.  
A function can return data as a result.  
In Python, a function is defined using the def keyword, followed by a function name and parentheses.  
To call a function, write its name followed by parentheses.  
Function names are case-sensitive (myFunction and myfunction are different).  

```python
def my_function():
  print("Hello")

my_function()
my_function()
my_function()
```

```python
# with arguments
def fahrenheit_to_celsius(fahrenheit):
  return (fahrenheit - 32) * 5 / 9
fahrenheit_to_celsius(95)
```

```python
def get_greeting():
  return "Hello from a function"

message = get_greeting()
message
```

print() displays text to the human user on the screen dosen't store the variable, while return passes a value back to the computer program to be stored in a variable or used in further calculations.

```python
def add_and_print(a, b):
    print(a + b)
result = add_and_print(7, 3) 
print(result)  
```

```python
def add_and_return(a, b):
    return a + b
result = add_and_return(7, 3)
total = result * 2  
print(total)
```

```python
#default parameter
def my_function(country = "Norway"):
  print("I am from", country)

my_function("Sweden")
my_function("India")
my_function()
my_function("Brazil")
```

```python
def my_function(x, y):
  return x + y

result = my_function(5, 3)
print(result)
```

```python
def my_function():
  return ["apple", "banana", "cherry"]

fruits = my_function()
print(fruits[0])
print(fruits[1])
print(fruits[2])
```

```python
def my_function(*kids):
  print("The youngest child is " + kids[2])

my_function("Emil", "Tobias", "Linus")
```

```python
def my_function(*numbers):
  if len(numbers) == 0:
    return None
  max_num = numbers[0]
  for num in numbers:
    if num > max_num:
      max_num = num
  return max_num
my_function(3, 7, 2, 9, 1)
```

## Generators  
Generators are functions that can pause and resume their execution.  
When a generator function is called, it returns a generator object, which is an iterator.  
The code inside the function is not executed yet, it is only compiled. The function only executes when you iterate over the generator.  
It is iterator itself written like a fun and uses yeild in place of return.  
Return ends the function but yield just pauses it.  

```python
def count_up_to(n):
  count = 1
  while count <= n:
    yield count
    count += 1

for num in count_up_to(5):
  print(num)
```

```python
#using Iter with generator
def large_sequence(n):
  for i in range(n):
    yield i
gen = large_sequence(10)
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
def fibonacci():
  a, b = 1, 1
  while True:
    yield a
    a, b = b, a + b
gen = fibonacci()
for i in range(10):
  print(next(gen))
```

```python
def squares(*numbers):
    for i in numbers:
        i= i**2
        yield i
numbers = squares(1,3,5,7,9)
for i in numbers:
    print(i)
```

```python
def gencubes(n):
    for num in range(n):
        yield num**3
gen = gencubes(10)
for i in gen:
    print(i)
```

```python
numbers = squares(2,4,6,8,10)
for i in numbers:
    print(i)
```

```python
# Iter() in python
numbers = [10, 20, 30, 40]

it = iter(numbers)

print(next(it))
print(next(it))
print(next(it))
print(next(it))
```

```python
def count():
    yield 10
    yield 20
    yield 30
gen = count()
iter = iter(gen)
print(next(iter))
print(next(iter))
print(next(iter))
```

### Lambda function  
A lambda function can take any number of arguments, but can only have one expression.  
Lambda is used as temporary function.  


```python
x = lambda a : a + 10
x(7)
```

```python
x = lambda a, b, c : a + b + c
x(5, 6, 2)
```

```python
# using in functions
def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(2)
mydoubler(11)
```

### map() , reduce() and filter() methods  
map() applies a function to each element in an iterable and returns a map object (iterator)  
map(function, iterable)  
  
reduce() repeatedly applies a function to the elements and reduces them into a single aggregated value.  
reduce(function, iterable)  

filter() is used to select elements from an iterable based on a condition. It returns only the elements for which the function returns True.  
filter(function, iterable)

```python
numbers = [1, 2, 3, 4, 5]
def square(x):
    return x ** 2
result = map(square, numbers)
list(result)
```

```python
result = map(lambda x: x **2, numbers)
list(result)
```

```python
a = [1, 2, 3]
b = [10, 20, 30]
result = map(lambda x, y: x + y, a, b)
list(result)
```

```python
# sum of numbers using reduce
from functools import reduce
numbers = [1, 2, 3, 4, 5]
result = reduce(lambda x, y: x + y, numbers)
result
```

```python
# finding max num
numbers = [10, 5, 20, 8, 15]
result = reduce(lambda x, y: x if x > y else y, numbers)
result
```

```python
# multipy the numbers
numbers = [1, 2, 3, 4, 5]
result = reduce(lambda x, y: x * y, numbers)
result
```

```python
# Using filter to return even numbers from list
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
list(even_numbers)
```

```python
result = filter(lambda x: x > 5, numbers)
list(result)
```

```python
# prints names starting with A
names = ["Alex", "Bob", "Alice", "David", "Andrew"]
result = filter(lambda name: name.startswith("A"), names)
list(result)
```

```python
def check_age(age):
    return age >= 18
ages = [12, 18, 25, 15, 30]
adults = filter(check_age, ages)
list(adults)
```
