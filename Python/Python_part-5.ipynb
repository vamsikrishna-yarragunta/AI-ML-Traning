# File Handling In Python  
"r" - Read - Default value. Opens a file for reading, error if the file does not exist  
"a" - Append - Opens a file for appending, creates the file if it does not exist   
"w" - Write - Opens a file for writing, creates the file if it does not exist  
"x" - Create - Creates the specified file, returns an error if the file exists

## Error Handling in Python  
The try block lets you test a block of code for errors.  
The except block lets you handle the error.  
The else block lets you execute code when there is no error.  
The finally block lets you execute code, regardless of the result of the try- and except blocks.  

```python
try:
    f=open("testfile1.txt","r")
except:
    print("The file dosen't exist")
else:
    print(f.read())
```

```python
def askint():
    try:
        val=int(input("Enter an Integer:"))
    except:
        print(" looks like you didnot enter integer!")
        try:
            val= int(input("Try again please enter integer"))
        except:
            print("handle it")
    finally:
        print("Finally, I excecuted!")
askint()
```

```python
try:
    f = open("sample.txt","r")
except:
    print("Can't read the file")
else: 
    print(f.read())
finally:
    print("Executed")
```

```python
try:
    f = open("sample.txt","w")
    f.write("Using error handling")
except:
    print("Can't read the file")
else: 
    print("sucessfully Updated the sample.txt file")
    f.close()
finally:
    print("Executed")
```

## Database Connectivity  
Database connectivity means creating a connection between a Python program and a database to perform operations like inserting, retrieving, updating, and deleting data.  

```python
import sqlite3
conn = sqlite3.connect("student.db")
print("Database Connected")
conn.close()
```

```python
conn = sqlite3.connect("student.db")
cursor = conn.cursor()
cursor.execute(" create table student_details( Id, Name, age)")
conn.commit()
conn.close()
```

```python
conn = sqlite3.connect("student.db")
cursor = conn.cursor()
cursor.execute("""
insert into student_details (Id, name, age)
VALUES (1, 'Ravi', 20),(2, 'Raju', 21),(3, 'Rakesh', 24)
""")
conn.commit()
conn.close()
```

```python
conn = sqlite3.connect("student.db")
cursor = conn.cursor()
cursor.execute("select* from student_details")
records = cursor.fetchall()
for row in records:
    print(row)
```

```python
cursor.execute(" Update student_details set name = 'Hari' where id = 2")
cursor.execute("select* from student_details")
records = cursor.fetchall()
for row in records:
    print(row)
```

```python
cursor.execute("delete from student_details where id=3")
cursor.execute("select* from student_details")
records = cursor.fetchall()
for row in records:
    print(row)
conn.close()
```
