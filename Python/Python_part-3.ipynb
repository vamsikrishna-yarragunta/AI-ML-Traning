### Slicing Of strings  
You can return a range of characters by using the slice syntax.  
Specify the start index,the end index and also the step size(optional), separated by a colon, to return a part of the string.  

```python
# String Slicing
b = "Hello, World!"
print(b[2:5])
```

```python
#slice from start
b = "Hello, World!"
print(b[:5])
```

```python
# slice to end
b = "Hello, World!"
print(b[2:])
```

```python
# slicing from back(negative indexing)
b = "Hello, World!"
print(b[-5:-2])
```

```python
# get all
b = "Hello, World!"
print(b[:])
```

```python
#reverse string
b = "Hello, World!"
print(b[::-1])
```

```python
# using step
b = "Hello, World!"
print(b[0:10:2])
```

```python
# use * to multiply
s = 'v'
print(50*s)
```

```python
keys = ['name', 'role', 'language']
values = ['Alice', 'Dev', 'Python']
user_profile = {k: v for k, v in zip(keys, values)}
user_profile
```

```python
# Shallow copy
import copy
student = {"name": "Emma",
           "marks": [85, 90, 95]}
print("Original befor copy:", student)
copied_student = copy.copy(student)
copied_student["marks"][1] = 100
copied_student["name"] = "alice"
print("Original after copy:", student)
print("Copy:", copied_student)
# Top-level changes will not affect the original, but nested updates will sync between both
```

```python
original = {"user": "Alice", "skills": ["Python", "SQL"]}
shallow_copied = original.copy()
shallow_copied["user"] = "Bob"
print(original["user"])      
shallow_copied["skills"].append("Docker")
original["skills"]
```

```python
# Deep copy
student = {"name": "Emma",
           "marks": [85, 90, 95]}

copied_student = copy.deepcopy(student)
copied_student["marks"][1] = 100
print("Original:", student)
print("Copy:", copied_student)
```
